```python
def perceptron(X,y): # yha input data pass ho rha h parameter m
    
    X = np.insert(X,0,1,axis=1) # yha hum x0 column add kr rhe h which will have value 1 in entire column
    weights = np.ones(X.shape[1])
    lr = 0.1 # set learning rate as 0.1
    
    for i in range(1000): # epoch = 1000 run loop 1000 times
        j = np.random.randint(0,100) # pick random obs from 100 obs
        y_hat = step(np.dot(X[j],weights)) # calculate predicted value (random obs dot product weights) w0.x0 + w1.x1 + w2.x2 and pass it to step function
        # model algorithm:
        # if val of y[j]-y_hat = 0, then point is correctly classified, update new weight same as old weight no need of any changes
        # if val of y[j]-y_hat = 1 or -1, then point is incorrectly classified, update new weight
        weights = weights + lr*(y[j]-y_hat)*X[j] 
        
    return weights[0],weights[1:] # return weights[0] => intercept and weights[1:] => coeff
        
```

```python
def step(z):
    return 1 if z>0 else 0
```

```python
intercept_,coef_ = perceptron(X,y)
```

```python
print(coef_) # coef_[0]=w1=1.30009809, coef_[1]=w2=0.08165687
print(intercept_) # w0=0.9
```

```python
# to plot line on graph to separate two classes, we need line equation
# y = m * x + b

# Ax + By + C = 0 => w1.x1 + w2.x2 + w0 = 0
# A, B, C => w1, w2, w0

# m => slope = -(A/B)
# b => Intercept = -(C/B)

m = -(coef_[0]/coef_[1]) # -(w1/w2)
b = -(intercept_/coef_[1]) # -(w0/w2)
```

# Animation: How perceptron is trained

```python
%matplotlib notebook
from matplotlib.animation import FuncAnimation
import matplotlib.animation as animation
```

```python
def perceptron(X,y):

    m = []
    b = []
    
    X = np.insert(X,0,1,axis=1)
    weights = np.ones(X.shape[1])
    lr = 0.1
    
    for i in range(200):
        j = np.random.randint(0,100)
        y_hat = step(np.dot(X[j],weights))
        weights = weights + lr*(y[j]-y_hat)*X[j] 
        
        m.append(-(weights[1]/weights[2]))
        b.append(-(weights[0]/weights[2]))

    return m,b
        
```

```python
m,b = perceptron(X,y)
```

```python
fig, ax = plt.subplots(figsize=(10,6))
x_i = np.arange(-3, 3, 0.1)
y_i = x_i * m[0] + b[0]
ax.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=100)
line, = ax.plot(x_i,y_i,'r-',linewidth=2)
plt.ylim(-3,3)


def update(i):
  label = 'epoch {0}'.format(i + 1)
  line.set_ydata(x_i*m[i] + b[i])
  ax.set_xlabel(label)


anim = FuncAnimation(fig, update, repeat=True, frames=200, interval=100)
```
